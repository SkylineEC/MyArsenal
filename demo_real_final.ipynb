## SSD Forward Pass
### Now just wrap the image in a Variable so it is recognized by PyTorch autograd

```python
from data import VOC_CLASSES as labels

def get_center(length,height,i,pt0,pt1,pt2,pt3):
    center = (0,0)
    if(i == 0):
        center = ((pt2+pt0)/2,(pt3+pt1)/2)
    if(i == 1):
        center = (length/4+(pt2+pt0)/2,(pt3+pt1)/2)
    if(i == 2):
        center = (length/2+(pt2+pt0)/2,(pt3+pt1)/2)
    if(i == 3):
        center = (length*3/4+(pt2+pt0)/2,(pt3+pt1)/2)
    if(i == 4):
        center = ((pt2+pt0)/2,height/3+(pt3+pt1)/2)
    if(i == 5):
        center = (length/4+(pt2+pt0)/2,height/3+(pt3+pt1)/2)
    if(i == 6):
        center = (length/2+(pt2+pt0)/2,height/3+(pt3+pt1)/2)
    if(i == 7):
        center = (length*3/4+(pt2+pt0)/2,height/3+(pt3+pt1)/2)
    if(i == 8):
        center = ((pt2+pt0)/2,height*2/3+(pt3+pt1)/2)
    if(i == 9):
        center = (length/4+(pt2+pt0)/2,height*2/3+(pt3+pt1)/2)
    if(i == 10):
        center = (length/2+(pt2+pt0)/2,height*2/3+(pt3+pt1)/2)
    if(i == 11):
        center = (length*3/4+(pt2+pt0)/2,height*2/3+(pt3+pt1)/2)
    if(i == 12):
        center = (length*1/8+(pt2+pt0)/2,height*1/6+(pt3+pt1)/2)
    if(i == 13):
        center = (length*3/8+(pt2+pt0)/2,height*1/6+(pt3+pt1)/2)
    if(i == 14):
        center = (length*5/8+(pt2+pt0)/2,height*1/6+(pt3+pt1)/2)
    if(i == 15):
        center = (length*1/8+(pt2+pt0)/2,height*1/2+(pt3+pt1)/2)
    if(i == 16):
        center = (length*3/8+(pt2+pt0)/2,height*1/2+(pt3+pt1)/2)
    if(i == 17):
        center = (length*5/8+(pt2+pt0)/2,height*1/2+(pt3+pt1)/2)
        
    if(i == 18):
        center = (length*1/4+(pt2+pt0)/2,height*1/6+(pt3+pt1)/2)
    if(i == 19):
        center = (length*1/2+(pt2+pt0)/2,height*1/6+(pt3+pt1)/2)
    if(i == 20):
        center = (length*1/4+(pt2+pt0)/2,height*1/2+(pt3+pt1)/2)
    if(i == 21):
        center = (length*1/2+(pt2+pt0)/2,height*1/2+(pt3+pt1)/2)
    if(i == 22):
        center = (length*3/8+(pt2+pt0)/2,height*1/3+(pt3+pt1)/2)
    return center
num_yslxb = 0
num_zylxb = 0
num_zhylxb = 0
num_wylxb = 0
num_gzlxb = 0
num_fylxb = 0
top_k=10
centerset = []*1000
dict = {'0':upper_1,'1':upper_2,'2':upper_3,'3':upper_4,
        '4':upper_5,'5':upper_6,'6':upper_7,'7':upper_8,
        '8':upper_9,'9':upper_10,'10':upper_11,'11':upper_12,
        '12':edge_1,'13':edge_2,'14':edge_3,'15':edge_4,
        '16':edge_5,'17':edge_6,'18':qua_1,'19':qua_2,'20':qua_3,
        '21':qua_4,'22':_center }

for image in range(23):
    xx = Variable(dict[str(image)].unsqueeze(0))     # wrap tensor in Variable
    if torch.cuda.is_available():
        xx = xx.cuda()
    y = net(xx)
    plt.figure(figsize=(20,20))
    colors = plt.cm.hsv(np.linspace(0, 1, 10)).tolist()
    plt.subplot(3,8,image+1)
    plt.subplots_adjust(wspace = 0)
    plt.imshow(rgb_images[image])  # plot the image for matplotlib
   
    currentAxis = plt.gca()

    detections = y.data
    # scale each detection back up to the image
    scale = torch.Tensor(rgb_images[image].shape[1::-1]).repeat(2)
    for i in range(detections.size(1)):
        flag = 0
        j = 0
        center = ()
        while detections[0,i,j,0] >= 0.6:
            score = detections[0,i,j,0]
            label_name = labels[i-1]
            
            
            #display_txt = '%s: %.2f'%(label_name, score)
            display_txt = (label_name)
            pt = (detections[0,i,j,1:]*scale).cpu().numpy()
            center = get_center(4032,3024,image,pt[0],pt[1],pt[2],pt[3])
            #print(center)
            #print(pt)
            coords = (pt[0], pt[1]), pt[2]-pt[0]+1, pt[3]-pt[1]+1#第一个参数是矩形左上角点坐标 tuple 
                                          #(x,y)第二个矩形长,第三个参数是矩形高
            center_small = ((pt[2]+pt[0])/2,(pt[3]+pt[1])/2),10,10
            for f in range(len(centerset)):
                #print(h)
                #print(center[0])
                #print(center[1])
                if image>=12 and (center[0]>=int(centerset[f][0]-100) and center[0]<=int(centerset[f][0]+100)) and (center[1]>=int(centerset[f][1]-100) and center[1]<=int(centerset[f][1]+100)):
                    flag = 1
                    break
                if (pt[2]-pt[0]+1)/(pt[2]-pt[0]+1)>= 1.3 or (pt[2]-pt[0]+1)/(pt[2]-pt[0]+1)<=0.77:
                    flag = 1
                    break
            if flag == 0:
                center_txt = '%.2f: %.2f'%center
                if label_name == 'yslxb':
                    num_yslxb += 1
                elif label_name == 'zylxb':
                    num_zylxb += 1
                elif label_name == 'zhylxb':
                    num_zhylxb += 1
                elif label_name == 'wylxb':
                    num_wylxb += 1
                elif label_name == 'gzlxb':
                    num_gzlxb += 1
                elif label_name == 'fylxb':
                    num_fylxb += 1                                                                                                  
                centerset.append(center)       
                color = colors[i]   
                currentAxis.add_patch(plt.Rectangle(*coords, fill=False, edgecolor=color, linewidth=1))
                currentAxis.add_patch(plt.Rectangle(*center_small, fill=False, edgecolor=color, linewidth=1))
                #currentAxis.text((pt[2]+pt[0])/2, (pt[3]+pt[1])/2, center_txt, bbox={'facecolor':color, 'alpha':0.2})
                currentAxis.text((pt[2]), pt[1], display_txt, bbox={'facecolor':color, 'alpha':0.1})
            j+=1 
#plt.show()
#print(centerset)

print("\033[1;36m---------------")
print("|原始粒细胞:",num_yslxb," |")
print("---------------")
print("|早幼粒细胞:",num_zylxb,"|")
print("---------------")
print("|中幼粒细胞:",num_zhylxb,"|")
print("---------------")
print("|晚幼粒细胞:",num_wylxb,"|")
print("---------------")
print("|杆状粒细胞:",num_gzlxb,"|")
print("---------------")
print("|分叶粒细胞:",num_fylxb,"|")
print("---------------\033[0m")
```
